```python
import pandas as pd
import numpy as np
import ast
import sklearn
import nltk
import pickle
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def converting_to_list(obj):
    G =[]
    for i in ast.literal_eval(obj):
        G.append(i['name'])
    return G
```

```python
movies['genres'] = movies['genres'].apply(converting_to_list)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(converting_to_list)
```

```python
def converting(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(converting)
```

```python
movies.head()
```

```python
def fetch_dir(obj):
    D =[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            D.append(i['name'])
            break
    return D
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
films = movies[['movie_id','title','tags']]
films
```

```python
films['tags'] = films['tags'].apply(lambda x:" ".join(x))
```

```python
films['tags'][0]
```

```python
films['tags'] = films['tags'].apply(lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    a = []
    for i in text.split():
        a.append(ps.stem(i))
    return " ".join(a)
```

```python
films['tags'] = films['tags'].apply(stem)
```

```python
films.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(films['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn .metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similar[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    m_index = films[films['title'] == movie].index[0]
    distance = similar[m_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x:x[1])[1:6]
    for i in movie_list:
        print(films.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
pickle.dump(films,open('movies.pkl','wb'))
```

```python
pickle.dump(films.to_dict(),open('movies_dict.pkl','wb'))
```
